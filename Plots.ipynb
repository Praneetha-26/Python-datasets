```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('spotify.csv')
df
```

```python
#Histogram
column='tempo'
plt.figure(figsize=(10,5))
sns.histplot(df[column],bins=30,kde=True,color='blue')
plt.title('Histogram')
plt.xlabel(column)
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Boxplot
column='tempo'
plt.figure(figsize=(6, 4))
sns.boxplot(x=df[column], color='blue')
plt.title('Boxplot')
plt.xlabel(column)
plt.grid(True)
plt.show()
```

```python
#KDE plot
column = 'tempo'
plt.figure(figsize=(8, 6))
sns.kdeplot(df[column], shade=True, color='blue')
plt.title('KDE Plot ')
plt.xlabel(column)
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='popularity', y='tempo', data=df)
plt.title('Scatterplot')
plt.show()
```
