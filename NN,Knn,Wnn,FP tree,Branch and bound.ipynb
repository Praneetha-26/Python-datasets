```python
import math
from collections import Counter
data = [
    {'point': (1,1.5), 'class': '1'},
    {'point': (1, 2.5), 'class': '1'},
    {'point': (1, 3.5), 'class': '1'},
    {'point': (2, 1.5), 'class': '1'},
    {'point': (2, 2.5), 'class': '1'},
    {'point': (2,3.5), 'class': '1'},
    {'point': (2, 4), 'class': '1'},
    {'point': (2.5, 2.5), 'class': '1'},
    {'point': (3.5, 1.5), 'class': '1'},
    {'point': (3.5, 2.5), 'class': '1'},
    {'point': (3.5,3.5), 'class': '2'},
    {'point': (3.5, 4.5), 'class': '2'},
    {'point': (4.5, 1.5), 'class': '2'},
    {'point': (4.5, 2.5), 'class': '2'},
    {'point': (4.5, 3.5), 'class': '2'},
    {'point': (5,4.5), 'class': '2'},
    {'point': (5, 5.5), 'class': '2'},
    {'point': (6, 3.5), 'class': '2'},
    {'point': (6, 4.5), 'class': '2'},
    {'point': (6, 5.5), 'class': '2'}
]
new_point = (2.6, 5.5)
K = 3
def euclidean(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
distances = sorted(data, key=lambda item: euclidean(item['point'], new_point))
k_nearest = distances[:K]
k_classes = [neighbor['class'] for neighbor in k_nearest]
predicted_class = Counter(k_classes).most_common(1)[0][0]
print(f"New point: {new_point}")
print(f"K={K} Nearest Neighbors: {[item['point'] for item in k_nearest]}")
print(f"Predicted Class: {predicted_class}")
```

```python
import math
data = [ 
    {'point': (1,1.5), 'class': '1'},
    {'point': (1, 2.5), 'class': '1'},
    {'point': (1, 3.5), 'class': '1'},
    {'point': (2, 1.5), 'class': '1'},
    {'point': (2, 2.5), 'class': '1'},
    {'point': (2,3.5), 'class': '1'},
    {'point': (2, 4), 'class': '1'},
    {'point': (2.5, 2.5), 'class': '1'},
    {'point': (3.5, 1.5), 'class': '1'},
    {'point': (3.5, 2.5), 'class': '1'},
    {'point': (3.5,3.5), 'class': '2'},
    {'point': (3.5, 4.5), 'class': '2'},
    {'point': (4.5, 1.5), 'class': '2'},
    {'point': (4.5, 2.5), 'class': '2'},
    {'point': (4.5, 3.5), 'class': '2'},
    {'point': (5,4.5), 'class': '2'},
    {'point': (5, 5.5), 'class': '2'},
    {'point': (6, 3.5), 'class': '2'},
    {'point': (6, 4.5), 'class': '2'},
    {'point': (6, 5.5), 'class': '2'}
]
new_point = (2.6, 5.5)
k =3
def euclidean(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
def weighted_knn(data, new_point, k):
    distances = []
    for item in data:
        dist = euclidean(item['point'], new_point)
        distances.append((dist, item['class']))
    distances.sort(key=lambda x: x[0])
    k_nearest = distances[:k]
    class_weights = {}
    for dist, cls in k_nearest:
        weight = 1 / (dist + 1e-5)  
        if cls in class_weights:
            class_weights[cls] += weight
        else:
            class_weights[cls] = weight
    predicted_class = max(class_weights.items(), key=lambda x: x[1])[0]
    return predicted_class, class_weights
predicted_class, weights = weighted_knn(data, new_point, k)
print(f"New point: {new_point}")
print(f"Predicted class: {predicted_class}")
print("Class weights:")
for cls, weight in weights.items():
    print(f"  Class {cls}: {weight:.4f}")
```

```python

transactions_by_class = {}
for item in data:
    cls = item['class']
    point_str = f"{item['point'][0]}_{item['point'][1]}"
    if cls not in transactions_by_class:
        transactions_by_class[cls] = []
    transactions_by_class[cls].append(point_str)
transaction_list = list(transactions_by_class.values())
class TreeNode:
    def __init__(self, item, count, parent):
        self.item = item
        self.count = count
        self.parent = parent
        self.children = {}
        self.link = None 
root = TreeNode(None, 1, None)
header_table = {}
for transaction in transaction_list:
    transaction = sorted(transaction)  
    current_node = root
    for item in transaction:
        if item in current_node.children:
            current_node.children[item].count += 1
        else:
            new_node = TreeNode(item, 1, current_node)
            current_node.children[item] = new_node
            if item not in header_table:
                header_table[item] = new_node
            else:
                node = header_table[item]
                while node.link is not None:
                    node = node.link
                node.link = new_node

        current_node = current_node.children[item]
def display_tree(node, indent=0):
    for child in node.children.values():
        print('  ' * indent + f"{child.item} ({child.count})")
        display_tree(child, indent + 1)

display_tree(root)
```

```python
max_total_distance=10
items = []
for item in data:
    pt = item['point']
    dist = math.sqrt((pt[0] - new_point[0])**2 + (pt[1] - new_point[1])**2)
    value = 1 / (dist + 1e-5)  # Avoid division by zero
    items.append({'point': pt, 'class': item['class'], 'value': value, 'cost': dist})

items.sort(key=lambda x: x['value'] / x['cost'], reverse=True)

class Node:
    def __init__(self, level, value, cost, bound, selected):
        self.level = level
        self.value = value
        self.cost = cost
        self.bound = bound
        self.selected = selected
def calculate_bound(node):
    if node.cost >= max_total_distance:
        return 0
    total_value = node.value
    total_cost = node.cost
    i = node.level
    while i < len(items) and total_cost + items[i]['cost'] <= max_total_distance:
        total_cost += items[i]['cost']
        total_value += items[i]['value']
        i += 1
    if i < len(items):
        total_value += (max_total_distance - total_cost) * items[i]['value'] / items[i]['cost']
    return total_value
queue = []
root = Node(0, 0, 0, 0.0, [])
root.bound = calculate_bound(root)
queue.append(root)

best_value = 0
best_selection = []

while queue:
    queue.sort(key=lambda x: x.bound, reverse=True)
    node = queue.pop(0)

    if node.bound > best_value and node.level < len(items):
    
        next_cost = node.cost + items[node.level]['cost']
        next_value = node.value + items[node.level]['value']
        next_selected = node.selected + [items[node.level]['point']]

        if next_cost <= max_total_distance and next_value > best_value:
            best_value = next_value
            best_selection = next_selected

        include_node = Node(node.level + 1, next_value, next_cost, 0.0, next_selected)
        include_node.bound = calculate_bound(include_node)
        if include_node.bound > best_value:
            queue.append(include_node)


        exclude_node = Node(node.level + 1, node.value, node.cost, 0.0, node.selected)
        exclude_node.bound = calculate_bound(exclude_node)
        if exclude_node.bound > best_value:
            queue.append(exclude_node)

# Step 5: Output
print("New point:", new_point)
print("Best total value:", round(best_value, 4))
print("Selected points:")
for pt in best_selection:
    print(" ", pt)
```
